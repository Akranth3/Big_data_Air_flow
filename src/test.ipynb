```python
import pandas as pd
import pickle
import sys

sys.path.append('../')

# Load the data
file_name = 'data.csv'

df = pd.read_csv('Data/' + file_name, low_memory=False)
```

```python
monthly_column_names = [col for col in df.columns if col[0:7] == 'Monthly']
monthly_column_names.append('DATE')

monthly_data = df[monthly_column_names]
monthly_data['MONTH'] = monthly_data['DATE'].apply(lambda x: x[5:7])

# Drop the date
monthly_data.drop('DATE', axis=1, inplace=True)
```

```python
monthly_data.isnull().sum()
```

```python
monthly_data.dropna(axis=1, how='all', inplace=True)
```

```python
day_column_names = [col for col in df.columns if col[0:5] == 'Daily']
day_column_names.append('DATE')

day_data = df[day_column_names]

#converting the date to month
day_data['MONTH'] = day_data['DATE'].apply(lambda x: x[5:7])
day_data.drop('DATE', axis=1, inplace=True)
```
